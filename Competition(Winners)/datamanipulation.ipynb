```python
def manipulate_vehiclemodel(data):
    
    # to lower case all to ease the manipulation
    data["VehicleModel"] = data["VehicleModel"].str.lower()
    
    # to remove some unnecessary words
    to_remove = ["i have the fastest model ever.. yes you got it right it's ", "my model which is ", " is the best", "i own " ]
    for i in to_remove:
        data["VehicleModel"] = data["VehicleModel"].str.replace(i, "")
    
    # for removing trailing dots
    data["VehicleModel"] = data["VehicleModel"].str.replace('.', "")
    
    # for removing non-ascii characters
    data["VehicleModel"] = data["VehicleModel"].str.replace(r'[^\x00-\x7F]', '', regex=True).str.strip()
    
    
    """
    This part was to make the model just one word for each group of brand, for example if the model was "prius v" it will be just "prius"
    so by that way we can reduce the number of unique values in the model column and make it more consistent as I thought.
    """
    brands = data['Brand'].unique()
    for brand in brands:
        mask = data['Brand'] == brand
        data.loc[mask, 'VehicleModel'] = data.loc[mask, 'VehicleModel'].astype(str)
        
        # process 'VehicleModel' column to keep only the first word if it contains more than one word
        data.loc[mask, 'VehicleModel'] = data.loc[mask, 'VehicleModel'].apply(lambda x: \
                                                                            ''.join(str(x).split(" ")[0]) if len(str(x).split(" ")) > 1 \
                                                                                                            else x)
    return data
```

Now let's try to fill the missing values in the model column.
- We will do that by following these steps:
    1. First we will replace 'nan' strings values with real NaN values to fill it correctly.
    2. Second we will group the data by brand, cylinder count, and capacity as I saw that model of the car depends on these feature the most.
    3. Third we will fill the missing values in the model column by the most frequent value in the group.
    4. Finally, we will fill the remaining missing values with 'unknown' category.

```python
def fill_missing_model(data):
    
    data['VehicleModel'] = data['VehicleModel'].replace('nan', pd.NA)
    to_fill_model = data.groupby(['Brand', 'CylinderCount', 'capacity'])['VehicleModel']
    data['VehicleModel'] = to_fill_model.transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'unknown'))
    return data
```

Now let's do the magic:

```python
# this might not be the best way to do it but I didn't want to waste more time in this column
# this function will make the year reasonable and consistent

def check_years(year):
    if year == str(pd.NA):
        return year
    if year[0] == '3':
        return '2'+ year[1:]
    if year[0] == '2' and year[1] >='8': 
        return '19'+ year[2:]
    if year[0] == '2' and year[1] != '0':
        return '20'+ year[2:]
    elif year[0] == '2' and year[1] == '0':
        if year[2] <= '2':
            return '20'+ year[2:]
        else:
            return '19'+ year[2:]
    return year
```

```python

def manipulate_years(data):
    
    def extract_year(year_str):
        
        import re
        match = re.search(r'(\d{4})', year_str)
        if match:
            return match.group(1)
        else:
            match = re.search(r'(\d+)y', year_str)
            if match:
                return 2021 - int(match.group(1))
            else:
                return pd.NA


    data["ManufacturingYear"] = data["ManufacturingYear"].astype(str).apply(extract_year)
    data.ManufacturingYear = data.ManufacturingYear.apply(lambda x: check_years(str(x)))
    data.ManufacturingYear = pd.to_numeric(data.ManufacturingYear, errors='coerce')
    return data
```

```python
manipulated = manipulate_years(manipulated)

print(is_nan(manipulated.ManufacturingYear))

print(manipulated.ManufacturingYear.nunique())

manipulated.ManufacturingYear.unique()
```

To fill missing values in the manufacturing year column, we will do the same steps as we did in the model and brand columns.

```python
def fill_missing_year(data):
    # Group by 'Brand' and 'VehicleModel'
    to_fill_years = data.groupby(['Brand', 'VehicleModel'])['ManufacturingYear']
    
    # Fill missing values in 'ManufacturingYear' with the most frequent year within the group,
    # or the overall mode if the group mode is empty
    data['ManufacturingYear'] = to_fill_years.transform(
        lambda x: x.fillna(
            x.mode().iloc[0] if not x.mode().empty else data['ManufacturingYear'].mode().iloc[0]
        )
    )
    
    # Convert 'ManufacturingYear' to integer type
    data['ManufacturingYear'] = data['ManufacturingYear'].astype(int)
    
    # Convert 'ManufacturingYear' to datetime and then to period (year only)
    # this is not a necessary step but I just did it to make the data more consistent
    data['ManufacturingYear'] = pd.to_datetime(data['ManufacturingYear'], format='%Y').dt.to_period('Y')
    
    return data
```

```python

manipulated = fill_missing_year(manipulated)

print(manipulated['ManufacturingYear'].isna().sum())
```

```python
import pandas as pd

def fill_manipulate_missing_color(data):
    # Convert color to lowercase
    data['color'] = data['color'].str.lower()
    
    # Replace 'nan' with actual missing values
    data['color'].replace('nan', pd.NA, inplace=True)
    
    # Group by 'Brand', 'VehicleModel', and 'Type' and calculate the mode for 'color'
    to_fill_color = data.groupby(['Brand', 'VehicleModel', 'Type'])['color']
    
    # Fill missing values with the mode within each group
    data['color'] = to_fill_color.transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'unknown'))
    
    return data
```

```python
def fill_manipulate_fuel(data):
    # Convert fuel to lowercase
    data['fuel'] = data['fuel'].str.lower()
    
    # Replace 'nan' with actual missing values
    data['fuel'].replace('nan', pd.NA, inplace=True)
    
    # Group by 'Brand', 'VehicleModel', and 'Type' and calculate the mode for 'fuel'
    to_fill_fuel = data.groupby(['Brand', 'VehicleModel', 'Type'])['fuel']
    
    # Fill missing values with the mode within each group
    data['fuel'] = to_fill_fuel.transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'other'))
    data['fuel'] = data['fuel'].astype('category')
    
    return data
```

```python
def manipulate_airbags(data):
    
    data['airbags'] = data['airbags'].str.extract(r'(-?\d+)', expand=False)
    data['airbags'] = pd.to_numeric(data['airbags'], errors='coerce')
    
    for i in data['airbags'].unique():
        if i < 0:
            data['airbags'] = data['airbags'].replace(i, 0)
        if i > 12:
            data['airbags'] = data['airbags'].replace(i, 12)
    data['airbags'] = data['airbags'].astype(int)
    
    return data
```
